##The dataset contains all the information needed to summarize the story of IPL 2022 so far. So letâ€™s start by looking at the number of matches won by each team in IPL 2022:

```python
figure = px.bar(Ipl_data, x=Ipl_data["match_winner"],
            title="Number of Matches Won in IPL 2022")
figure.show()
```

##So, Gujrat is leading the tournament by winning eight matches. It is an achievement as a new team for Gujrat in IPL.

```python
fig = px.pie(Ipl_data,names="match_winner", title="Number of tosses Won in IPL 2022")
fig.show()
```

```python
figure = px.bar(Ipl_data, x=Ipl_data["toss_winner"],
            title="Number of tosses Won in IPL 2022")
figure.show()
```

```python
fig = px.pie(Ipl_data,names="toss_winner", title="Number of tosses Won in IPL 2022")
fig.show()
```

```python
figure = px.bar(Ipl_data, x=Ipl_data["top_scorer"],
            title="Top Scorers in IPL 2022")
figure.show()
```

##Here we can see Jos Buttler has been a top scorer in 5 matches. He is looking in great touch.

##**Letâ€™s analyze it deeply by including the runs scored by the top scorers:**

```python
figure = px.bar(Ipl_data, x=Ipl_data["top_scorer"], 
                y = Ipl_data["highscore"], 
                color = Ipl_data["highscore"],
            title="Top Scorers in IPL 2022")
figure.show()
```

##So Jos Buttler has scored three centuries, and KL Rahul has scored two centuries.

##**Now letâ€™s have a look at the most player of the match awards till now in IPL 2022:**

```python
figure = px.bar(Ipl_data, x = Ipl_data["player_of_the_match"], 
         title="Most Player of the Match Awards")
figure.show()
```

##So Kuldeep Yadav is leading in the list of players of the match awards with four matches. It is a great tournament for Kuldeep Yadav this year.

## **Now letâ€™s have a look at the bowlers with the best bowling figures in most of the matches:**

```python
figure = px.bar(Ipl_data, x=Ipl_data["best_bowling"],
            title="Best Bowlers in IPL 2022")
figure.show()
```

##You can see Yuzvendra Chahal having the best bowling figures in four matches. So this is a great tournament for Yuzvendra Chahal this year too.

#**Now letâ€™s have a look at whether most of the wickets fall while setting the target or while chasing the target:**

```python
figure = go.Figure()
figure.add_trace(go.Bar(
    x=Ipl_data["venue"],
    y=Ipl_data["first_ings_wkts"],
    name='First Innings Wickets',
    marker_color='darkorange'
))
figure.add_trace(go.Bar(
    x=Ipl_data["venue"],
    y=Ipl_data["second_ings_wkts"],
    name='Second Innings Wickets',
    marker_color='green'
))
figure.update_layout(barmode='group', xaxis_tickangle=-45)
figure.show()
```

##So in the Wankhede Stadium in Mumbai and MCA Stadium in Pune, most wickets fall while chasing the target. And in the other two stadiums, most wickets fall while setting the target. So this is how you can analyze and summarize the story of IPL 2022 using Python.
